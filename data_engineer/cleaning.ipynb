```python
import pandas as pd

df = pd.read_csv('../data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
missing = df.isnull().sum()
missing = missing[missing > 0]
if missing.empty:
    print("No missing values\n")
else:
    percent = (missing / len(df)) * 100
    result = pd.DataFrame({
        "MissingCount": missing,
        "MissingPercent": percent
    }).sort_values("MissingPercent", ascending=False)
    display(result)
```

```python
def analyze_df(df):
    for col in df.columns:
        print("="*50)
        print(f"Column: {col}")
        print(f"Type: {df[col].dtype}")

        print("Unique count:", df[col].nunique())

        print("\nTop values:")
        print(df[col].value_counts().head())

        if pd.api.types.is_numeric_dtype(df[col]):
            print("\nStats:")
            print(df[col].describe())

analyze_df(df)
```

## drop
EmployeeCount (constant = 1)
Over18 (constant = Y)
StandardHours (constant = 80)
EmployeeNumber (ID, không có ý nghĩa ML)

## encoding
Attrition (target)
BusinessTravel
Department
EducationField
Gender
JobRole
MaritalStatus
OverTime

```python
# drop
df = df.drop(columns=[
    "EmployeeCount",
    "Over18",
    "StandardHours",
    "EmployeeNumber"
])

# binary encoding
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})
df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})
df["OverTime"] = df["OverTime"].map({"Yes": 1, "No": 0})

# one hot encoding
categorical_cols = [
    "BusinessTravel",
    "Department",
    "EducationField",
    "JobRole",
    "MaritalStatus"
]

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print(df.info())
print(df.head())
```

```python
analyze_df(df)
```

```python
df.columns = (
    df.columns
      .str.replace(' ', '_')
      .str.replace('&', 'and')
      .str.replace('-', '_')
)
```

```python
df.to_csv('../data/cleaned/WA_Fn-UseC_-HR-Employee-Attrition.csv', index=False)
```
